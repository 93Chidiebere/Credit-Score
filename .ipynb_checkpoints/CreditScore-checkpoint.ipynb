```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Target distribution
ax1 = axes[0, 0]
train_df['SeriousDlqin2yrs'].value_counts().plot(kind='bar', ax=ax1, color=['green', 'red'])
ax1.set_title('Target Variable Distribution', fontsize=14, fontweight='bold')
ax1.set_xlabel('SeriousDlqin2yrs')
ax1.set_ylabel('Count')
ax1.set_xticklabels(['No Default (0)', 'Default (1)'], rotation=0)

# Age distribution
ax2 = axes[0, 1]
ax2.hist(train_df['age'], bins=50, color='skyblue', edgecolor='black')
ax2.set_title('Age Distribution', fontsize=14, fontweight='bold')
ax2.set_xlabel('Age')
ax2.set_ylabel('Frequency')

# Monthly Income distribution (log scale)
ax3 = axes[1, 0]
train_df['MonthlyIncome'].dropna().apply(np.log1p).hist(bins=50, ax=ax3, color='lightgreen', edgecolor='black')
ax3.set_title('Monthly Income Distribution (Log Scale)', fontsize=14, fontweight='bold')
ax3.set_xlabel('Log(Monthly Income)')
ax3.set_ylabel('Frequency')

# Debt Ratio distribution
ax4 = axes[1, 1]
train_df['DebtRatio'].clip(0, 2).hist(bins=50, ax=ax4, color='salmon', edgecolor='black')
ax4.set_title('Debt Ratio Distribution (Clipped at 2)', fontsize=14, fontweight='bold')
ax4.set_xlabel('Debt Ratio')
ax4.set_ylabel('Frequency')

plt.tight_layout()
plt.savefig('eda_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

# Correlation heatmap
print("\nGenerating correlation heatmap...")
plt.figure(figsize=(12, 10))
correlation = train_df.drop('Unnamed: 0', axis=1).corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
